```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import bernoulli, binom, poisson, uniform, norm, expon

sns.set(style="whitegrid")
np.random.seed(42)
```

```python
p = 0.3  # probability of success
n = 10000  # number of experiments

samples = bernoulli.rvs(p, size=n)
```

```python
sns.countplot(x=samples)
plt.xticks([0, 1], ['Failure (0)', 'Success (1)'])
plt.title("Bernoulli Distribution Samples")
plt.show()
```

```python
n_trials = 10
p = 0.3
n_experiments = 10000

samples = binom.rvs(n_trials, p, size=n_experiments)
```

```python
sns.histplot(samples, bins=range(0, n_trials+2), kde=False)
plt.title("Binomial Distribution (n=10, p=0.3)")
plt.xlabel("Number of successes")
plt.show()
```

```python
lam = 4  # average rate
n_samples = 10000

samples = poisson.rvs(lam, size=n_samples)
```

```python
sns.histplot(samples, bins=range(0, samples.max()+2), stat='probability')
plt.title("Poisson Distribution (=4)")
plt.xlabel("Number of events")
plt.show()
```

```python
a, b = 0, 10
samples = uniform.rvs(loc=a, scale=b-a, size=100000)
```

```python
sns.histplot(samples, bins=30, stat='density')
plt.title("Uniform Distribution U(0,10)")
plt.xlabel("Value")
plt.show()
```

```python
mu = 50
sigma = 10
samples = norm.rvs(mu, sigma, size=10000)
```

```python
sns.histplot(samples, bins=40, stat='density')
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 200)
plt.plot(x, norm.pdf(x, mu, sigma), color='red')
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.show()
```

```python
lam = 0.5  # rate
samples = expon.rvs(scale=1/lam, size=10000)
```

```python
sns.histplot(samples, bins=40, stat='density')
x = np.linspace(0, samples.max(), 200)
plt.plot(x, expon.pdf(x, scale=1/lam), color='red')
plt.title("Exponential Distribution")
plt.xlabel("Waiting time")
plt.show()
```

```python
# =====================================================
# LAW OF LARGE NUMBERS DEMO
# =====================================================

import numpy as np
import matplotlib.pyplot as plt


# -----------------------------------------------------
# 1. Choose a Non-Normal Distribution
# -----------------------------------------------------
# Exponential distribution (mean = 1/lambda)
lambda_param = 2
true_mean = 1 / lambda_param

n_samples = 5000
samples = np.random.exponential(scale=true_mean, size=n_samples)

# -----------------------------------------------------
# 2. Compute Running Mean
# -----------------------------------------------------

running_mean = np.cumsum(samples) / np.arange(1, n_samples + 1)

# -----------------------------------------------------
# 3. Plot Convergence
# -----------------------------------------------------

plt.figure(figsize=(8,5))
plt.plot(running_mean, label="Sample Mean")
plt.axhline(true_mean, color='red', linestyle='--', label="True Mean")
plt.xlabel("Number of Samples")
plt.ylabel("Running Average")
plt.title("Law of Large Numbers Demonstration")
plt.legend()
plt.show()

print("Final sample mean:", running_mean[-1])
print("True mean:", true_mean)
```
