```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
np.random.seed(42)

# -----------------------------------------------------
# 1. Create Realistic E-Commerce Dataset
# -----------------------------------------------------

n = 500

df = pd.DataFrame({
    "Age": np.random.normal(35, 10, n),
    "Annual_Income": np.random.normal(75, 20, n),
    "Spending_Score": np.random.normal(60, 15, n),
    "Gender": np.random.choice(["Male", "Female"], n),
    "Purchase_Category": np.random.choice(
        ["Electronics", "Fashion", "Home", "Beauty"], n
    )
})

df.head()

# -----------------------------------------------------
# 2. Univariate Visualizations
# -----------------------------------------------------

# Histogram
sns.histplot(df["Annual_Income"], kde=True)
plt.title("Distribution of Annual Income")
plt.show()

# Boxplot
sns.boxplot(x=df["Spending_Score"])
plt.title("Boxplot of Spending Score")
plt.show()

# Barplot (Categorical count)
sns.countplot(x="Purchase_Category", data=df)
plt.title("Category Distribution")
plt.show()

# -----------------------------------------------------
# 3. Bivariate Visualizations
# -----------------------------------------------------

# Scatterplot
sns.scatterplot(
    x="Annual_Income",
    y="Spending_Score",
    hue="Gender",
    data=df
)
plt.title("Income vs Spending")
plt.show()

# Boxplot with Category
sns.boxplot(
    x="Purchase_Category",
    y="Spending_Score",
    data=df
)
plt.title("Spending by Category")
plt.xticks(rotation=30)
plt.show()

# -----------------------------------------------------
# 4. Correlation Heatmap
# -----------------------------------------------------

corr = df[["Age","Annual_Income","Spending_Score"]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

# -----------------------------------------------------
# 5. Pairplot
# -----------------------------------------------------

sns.pairplot(df[["Age","Annual_Income","Spending_Score"]])
plt.show()
```
