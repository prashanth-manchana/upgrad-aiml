```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

lam = 0.2  # average 5 minutes waiting time
n = 5000

waiting_times = np.random.exponential(scale=1/lam, size=n)
```

```python
running_avg = np.cumsum(waiting_times) / np.arange(1, n+1)
```

```python
plt.figure(figsize=(10,4))
plt.plot(running_avg, label="Empirical Average")
plt.axhline(1/lam, color='red', linestyle='--', label="True Expected Value")
plt.xlabel("Number of days observed")
plt.ylabel("Average waiting time")
plt.title("Law of Large Numbers: Waiting Time Example")
plt.legend()
plt.show()
```

```python
import seaborn as sns

spend = np.random.exponential(scale=1000, size=10000)

sns.histplot(spend, bins=50)
plt.title("Original Spend Distribution (Highly Skewed)")
plt.xlabel("Spend")
plt.show()
```

```python
def sample_means(data, sample_size, n_samples=5000):
    means = []
    for _ in range(n_samples):
        sample = np.random.choice(data, size=sample_size, replace=True)
        means.append(sample.mean())
    return np.array(means)
```

```python
means_1 = sample_means(spend, sample_size=1)

sns.histplot(means_1, bins=40, stat='density')
plt.title("Sample Size = 1 (Original Distribution)")
plt.show()
```

```python
means_5 = sample_means(spend, sample_size=5)

sns.histplot(means_5, bins=40, stat='density')
plt.title("Sample Size = 5")
plt.show()
```

```python
means_30 = sample_means(spend, sample_size=30)

sns.histplot(means_30, bins=40, stat='density')
plt.title("Sample Size = 30 (CLT kicks in)")
plt.show()
```

```python
means_100 = sample_means(spend, sample_size=100)

sns.histplot(means_100, bins=40, stat='density')
plt.title("Sample Size = 100 (Very Normal)")
plt.show()
```
