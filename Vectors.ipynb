```python
def mat_vec_mul(A, x):
    return [
        A[0][0]*x[0] + A[0][1]*x[1],
        A[1][0]*x[0] + A[1][1]*x[1]
    ]

A = [[2, 0],
     [0, 1]]

x = [1, 3]
print(mat_vec_mul(A, x))
```

```python
u = [1, 2]
v = [3, 1]
c = 2

A = [[1, 2],
     [0, 1]]

left = mat_vec_mul(A, [u[0]+v[0], u[1]+v[1]])
right = [a+b for a,b in zip(mat_vec_mul(A,u), mat_vec_mul(A,v))]

print(left, right)
```

```python
import math

theta = math.pi/4
A = [
    [math.cos(theta), -math.sin(theta)],
    [math.sin(theta),  math.cos(theta)]
]

x = [1, 0]
print(mat_vec_mul(A, x))
```

```python
def inverse_2x2(A):
    a,b = A[0]
    c,d = A[1]
    det = a*d - b*c
    return [[ d/det, -b/det],
            [-c/det,  a/det]]

A = [[2, 1],
     [1, 1]]

A_inv = inverse_2x2(A)

x = [3, 1]
print(mat_vec_mul(A_inv, mat_vec_mul(A, x)))
```
