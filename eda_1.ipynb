```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Employee_Specifics.csv')
```

```python
plt.figure(figsize=(8,4), dpi=100)
sns.scatterplot(data=df, x='Rating', y='Compensation', alpha=0.5, color='blue')
plt.title('Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Compensation')
plt.show()
```

```python
###Rating Feature
plt.figure(figsize=(6,3), dpi=100)
sns.countplot(data=df, x='Rating', alpha=0.5, edgecolor='linen')
plt.title('Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
###Rating Feature
plt.figure(figsize=(6,3), dpi=100)
sns.histplot(data=df, x='Compensation', alpha=0.5, edgecolor='linen')
plt.title('Scatter Plot')
plt.xlabel('Compensation')
plt.ylabel('Count')
plt.show()
```
