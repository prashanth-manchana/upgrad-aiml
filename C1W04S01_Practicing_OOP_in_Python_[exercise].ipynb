## Implementing `__init__()` Constructor and Creating an Object

Now, we want to understand how to initialise objects with attributes when they are created. The `__init__()` constructor method will allow us to pass data when creating instances of a class, which is important for managing student data in our system.

Create a class named `Student` that includes an `__init__()` constructor to initialise attributes such as `name`, `age`, and `course`. Then, create an object of the `Student` class, pass values to the constructor, and display an introduction of the student.

```python
# Define the Student class with an __init__ constructor
class Student:
    def __init__(self, name, age, course):
        self.name = name
        self.age = age
        self.course = course

    def introduction(self):
        return f"Hello, I am {self.name}, a {self.age}-year-old student enrolled in {self.course}."
```

```python
# Create an object of the Student class
student1 = Student("Snehal", 25, "Data Science")
print(student1.name)
print(student1.course)
print(student1.introduction())
```

```python
student2 = Student("Ankit", 22, "Gen AI")
print(student2.name)
print(student2.course)
print(student2.introduction())
```

## Adding Instance Methods and Updating Object Attributes

Now, we are adding functionality to the Student class by introducing a method that allows the course of a student to be updated. This enables dynamic modification of student data, which is a core feature of a management system.

Add a method to the `Student` class that allows updating the student's course. Then, create an object of the `Student` class, display its initial course, update the course, and display the updated course.

```python
# Define the Student class with a method to update course
class Student:
    def __init__(self, name, age, course, location):
        self.name = name
        self.age = age
        self.course = course
        self.location = location

    def introduction(self):
        return f"Hello, I am {self.name}, from {self.location} a {self.age}-year-old student enrolled in {self.course}."

    def change_course(self, new_course):
        self.course = new_course  # Method to change the course
```

```python
# Create an object and update the course
student1 = Student("Rishab", 25, "Data Science","Bangalore")
print(student1.course)
print(student1.introduction())  # Output: Hello, I am John, a 21-year-old student enrolled in Data Science.
```

```python
# Changing the course
student1.change_course("Machine Learning")
print(student1.course)
print(student1.introduction())  # Output: Hello, I am John, a 21-year-old student enrolled in Machine Learning.
```

## Implementing Class Variables and Class Methods

Now, we practice method overriding, where the subclass (`GraduateStudent`) overrides a method (`introduction()`) from the parent class (`Student`). This allows the subclass to provide a specialised version of the method, which is essential for tailoring functionalities in a system like a student management system.

In the `Student` class, add a class variable `total_students` to track how many students have been created. Add a class method to retrieve the total number of students. Then, create multiple student objects and call the class method to get the total number of students.

```python
# Define the Student class with class variables
class Student:
    total_students = 0  # Class variable to track the total number of students

    def __init__(self, name, age, course):
        self.name = name
        self.age = age
        self.course = course
        Student.total_students += 1  # Increment total_students

    def introduction(self):
        return f"Hello, I am {self.name}, a {self.age}-year-old student enrolled in {self.course}."

    @classmethod
    def get_total_students(clss):
        return f"Total students enrolled: {clss.total_students}"
```

```python
# Create student objects
student1 = Student("Paras", 24, "Data Science")
student2 = Student("Phani", 24, "Machine Learning")
# Call the class method to get the total number of students
print(Student.get_total_students())
```

```python
# Add another student objects
student3 = Student("Alice", 22, "Machine Learning")

# Call the class method to get the total number of students
print(Student.get_total_students())
```

## Implementing Inheritance

We are practicing inheritance by creating a `GraduateStudent` class that extends the `Student` class. This allows us to reuse functionality from the base class and add additional features specific to graduate students.

Create a `GraduateStudent` class that inherits from the `Student` class. Add an additional attribute for `research_topic` and implement a method that describes the research. Then, create an object of the `GraduateStudent` class and display the student's introduction and research.

```python
# Define the Student and GraduateStudent classes
class Student:
    def __init__(self, name, age=21, course="Gen AI"):
        self.name = name
        self.age = age
        self.course = course


    def introduction(self):
        return f"Hello, I am {self.name}, a {self.age}-year-old student enrolled in {self.course}."

class GraduateStudent(Student):
    def __init__(self, name, age, course, research_topic):
        super().__init__(name, age, course)  # Call the parent class constructor
        self.research_topic = research_topic

    def research(self):
        return f"{self.name} is researching on {self.research_topic}."
```

```python
# Demonstrate that the Student class does not have a research_topic attribute
student1 = Student("Lavanya")
print(student1.introduction())  # Output: Hello, I am John, a 21-year-old student enrolled in Data Science.
```

```python
# Create a GraduateStudent object
grad_student1 = GraduateStudent("Puja", 24, "Gen AI", "AI for Healthcare")

# Print introduction for both students
print(grad_student1.introduction())  # Output: Hello, I am David, a 24-year-old student enrolled in Data Science.
print(grad_student1.research())  # Output: David is researching on AI for Healthcare.
```

## Method Overriding in Python

In this step, we override the `introduction()` method in the `GraduateStudent` class to include information about the research topic. This demonstrates how we can change the behavior of a method in a subclass.

Override the `introduction()` method in the `GraduateStudent` class to include information about the student's `research_topic`. Then, create an object of the `GraduateStudent` class and display the student's introduction using the overridden method.

```python
# Define the Student and GraduateStudent classes with overridden method
class Student:
    def __init__(self, name, age=23, course="AI"):
        self.name = name
        self.age = age
        self.course = course

    def introduction(self):
        return f"Hello, I am {self.name}, a student enrolled in {self.course}."

    def introduction(self,age):
        self.age = age
        return f"Hello, I am {self.name}, a {self.age}-year-old student enrolled in {self.course}."

class GraduateStudent(Student):
    def __init__(self, name, age, course, research_topic="AI"):
        super().__init__(name, age, course)
        self.research_topic = research_topic

    def introduction(self):
        return f"Hello, I am {self.name}, a graduate student researching on {self.research_topic} in {self.course}."
```

```python
student1 = Student("Animesh")
print(student1.introduction(22))
```

```python
# Creating a Student object and calling the introduction method to display basic student details
student1 = Student("Sreeja", 21, "Generative AI and its RoI at scale")
print(student1.introduction())
```

```python
# Create a GraduateStudent object and show the overridden introduction
grad_student2 = GraduateStudent("Puja", 25, "Data Science")
print(grad_student2.introduction())  # Output: Hello, I am Anna, a graduate student researching on Natural Language Processing in Data Science.
```
